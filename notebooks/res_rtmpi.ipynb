```python
import datetime as dt
import time
import sys, os, struct
sys.path.append('/davitpy')
from models import raydarn
import subprocess as subp
from os import path
```

```python
sTime = dt.datetime(2012, 11, 18, 5)
eTime = sTime + dt.timedelta(hours=2)
radar = 'bks'
```

```python
tt = time.time()
rto = raydarn.rtRun(sTime, rCode=radar, outDir='/tmp', debug=True, nprocs=4)
print (time.time()-tt)
```

```python
nprocs = 4
timeit = {}
for nprocs in range(1,5):
    tmp = []
    for i in range(5):
        tt = time.time()
        rto = raydarn.rtRun(sTime, eTime, rCode=radar, outDir='/tmp', nprocs=nprocs)
        tmp.append(time.time()-tt)
        print nprocs, i, tmp[-1]
        fname = path.join(rto.outDir, 'rays.{}.inp'.format(rto.fExt))
        subp.call(['rm', fname])
        fname = path.join(rto.outDir, 'ionos.{}.inp'.format(rto.fExt))
        subp.call(['rm', fname])
        fname = path.join(rto.outDir, 'edens.{}.inp'.format(rto.fExt))
        subp.call(['rm', fname])
        fname = path.join(rto.outDir, 'ranges.{}.inp'.format(rto.fExt))
        subp.call(['rm', fname])
    timeit[nprocs] = tmp
```
