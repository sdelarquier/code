```python
#draw a vector
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
from matplotlib.font_manager import FontProperties

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

```python
def plot3d(view_elev, view_azim, ax):
    # Main and Interf array
    O = [0,0,0]
    dsep = 9.
    C = [0,dsep*.75,0]
    
    # Bore sight
    a1 = Arrow3D([0,0],[O[1]-2,O[1]+dsep],[0,0], 
        mutation_scale=20, lw=1, arrowstyle="-|>", color="k", ls='solid')
    ax.add_artist(a1)
    
    # Beam direction
    theta = radians(90)
    a2 = Arrow3D([0,dsep*sin(theta)],[0,dsep*cos(theta)],[0,0], 
        mutation_scale=20, lw=1, arrowstyle="-|>", color="k", ls='dashed')
    
    # Elevation direction
    el = radians(20)
    ax.plot3D([0,8*cos(el)*sin(theta)],[0,8*cos(el)*cos(theta)],[0,8*sin(el)],'r--')
    ax.plot3D([0,0],[0,8*cos(el)],[0,8*sin(el)],'r-')
    ax.add_artist(a2)
    
    A = [dsep*cos(theta)*cos(el)*cos(el)*sin(theta), 
         dsep*cos(theta)*cos(el)*cos(el)*cos(theta),
         dsep*cos(theta)*cos(el)*sin(el)]
    B = [dsep*.75*sin(theta),dsep*.75*cos(theta),0]
    
    # Magnetic field
    dip = -radians(70)
    Blen = 5
    ax.plot3D([C[0],C[0]],
              [C[1],C[1]+Blen*cos(dip-pi)],
              [C[2],C[2]+Blen*sin(dip-pi)],'b-')
    ax.plot3D([B[0],B[0]],
              [B[1],B[1]+Blen*cos(dip-pi)],
              [B[2],B[2]+Blen*sin(dip-pi)],'b--')
    
    # Mark k vector
    ael1 = Arrow3D([8*cos(el)*sin(theta),6*cos(el)*sin(theta)],
        [8*cos(el)*cos(theta),6*cos(el)*cos(theta)],
        [8*sin(el),6*sin(el)], 
        mutation_scale=20, lw=1, arrowstyle="<|-", color="r", linestyle="dashed")
    ax.add_artist(ael1)
    ael2 = Arrow3D([8*cos(el)*sin(0),6*cos(el)*sin(0)],
        [8*cos(el)*cos(0),6*cos(el)*cos(0)],
        [8*sin(el),6*sin(el)], 
        mutation_scale=20, lw=1, arrowstyle="<|-", color="r")
    ax.add_artist(ael2)
    
    # Mark corner angles
    # At B first
    Bnorm = sqrt(B[0]**2 + B[1]**2 + B[2]**2)*4
    Bun = [b/Bnorm for b in B]
    BC = map(lambda b,c:c-b, B, C)
    BCnorm = sqrt(BC[0]**2 + BC[1]**2 + BC[2]**2)*4
    BCun = [bc/BCnorm for bc in BC]
    CBun = [-bc for bc in BCun]
    Bperp = map(sum,zip(B,Bun,BCun))
    BA = map(lambda b,a:a-b, B, A)
    BAnorm = sqrt(BA[0]**2 + BA[1]**2 + BA[2]**2)*4
    BAun = [ba/BAnorm for ba in BA]
    BAperp = map(sum,zip(B,BCun,BAun))
    # Then at A
    Anorm = sqrt(A[0]**2 + A[1]**2 + A[2]**2)*4
    Aun = [a/Anorm for a in A]
    ABun = [-ba for ba in BAun]
    Aperp = map(sum,zip(A,Aun,ABun))
    AC = map(lambda c,a:c-a, C, A)
    ACnorm = sqrt(AC[0]**2 + AC[1]**2 + AC[2]**2)*4
    ACun = [ac/ACnorm for ac in AC]
    Anun = [-a for a in Aun]
    ACperp = map(sum,zip(A,ACun,Anun))
    
    # Set view
    ax.view_init(view_elev, view_azim)
    #ax.view_init(90,0)
    ax.set_xlim3d([-1,6])
    ax.set_xlabel('X')
    ax.set_ylim3d([-1,6])
    ax.set_ylabel('Y')
    ax.set_zlim3d([-1,6])
    ax.set_zlabel('Z')
    rcParams.update({'font.size': 14})
    
    # Annotation
    dshift = 8
    ax.text(B[0]+.25*dsep,
            B[1],
            -.2,
            "East", 
            ha = 'left', va = 'top')
    ax.text(0,
            C[1]+.25*dsep,
            -.2,
            "North", 
            ha = 'left', va = 'top')
    dshift = 28
    ax.text(O[0]+dshift*Aun[0],O[1]+dshift*Aun[1],O[2]+dshift*Aun[2]+.2, 
        "k", 
        ha = 'left', va = 'bottom', 
        color='r', fontproperties = FontProperties(weight='bold'))
    ax.text(O[0],8*cos(el)*cos(0),8*sin(el), 
        "k", 
        ha = 'left', va = 'bottom', 
        color='r', fontproperties = FontProperties(weight='bold'))
    ax.text(O[0],C[1]+Blen*cos(dip-pi),C[2]+Blen*sin(dip-pi), 
        "B", 
        color='b', fontproperties = FontProperties(weight='bold'), 
        ha = 'left', va = 'center')
    ax.text(B[0],B[1]+Blen*cos(dip-pi),B[2]+Blen*sin(dip-pi), 
        "B", 
        color='b', fontproperties = FontProperties(weight='bold'), 
        ha = 'left', va = 'center')
    
    ax.set_axis_off()
    rcParams.update({'font.size':20})
```

```python
fig = figure(figsize=(10,30))

ax = fig.add_subplot(311, projection='3d')
plot3d(20, 30, ax)

ax = fig.add_subplot(312, projection='3d')
plot3d(0, 0, ax)

ax = fig.add_subplot(313, projection='3d')
plot3d(0, 90, ax)

fig.tight_layout()
savefig('/home/sebastien/Desktop/azimbias.3d.eps')
```
