```python
import shlex
```

```python
with open('iri0UT.lst.txt', 'r') as f:
    data = f.readlines()
alt = zeros(len(data))
nelNight = zeros(len(data))
for i,ldat in enumerate(data):
    ldat = shlex.split(ldat)
    if len(ldat) == 0: continue
    alt[i] = ldat[0]
    nelNight[i] = ldat[1]
with open('iri12UT.lst.txt', 'r') as f:
    data = f.readlines()
nelDay = zeros(len(data))
for i,ldat in enumerate(data):
    ldat = shlex.split(ldat)
    if len(ldat) == 0: continue
    nelDay[i] = ldat[1]

with open('msis.txt', 'r') as f:
    data = f.readlines()
altM = zeros(len(data))
T = zeros(len(data))
for i,ldat in enumerate(data):
    ldat = shlex.split(ldat)
    if len(ldat) == 0: continue
    altM[i] = ldat[0]
    T[i] = ldat[1]
```

```python
rcParams.update({'font.size': 15})
fig = figure(figsize=(14,9))

ax1 = fig.add_axes([0.06, 0.1, 0.49, 0.85])
p11, = ax1.plot(T,altM,'g')
ax1.set_yscale('log')
ax1.set_ylim([1,2e3])
ax1.yaxis.set_major_formatter(ScalarFormatter())
ax1.set_xlabel('Temperature [K]')
ax1.set_ylabel('Altitude [km]')
ax1.tick_params('x',pad=16)
xticks([0,200,400,600,800,1000])
xran = [0,1200]
ax1.text(400, 200, 'Thermosphere')
ax1.plot(xran,[100,100],'k--')
ax1.text(400, 70, 'Mesosphere')
ax1.plot(xran,[45,45],'k--')
ax1.text(400, 25, 'Stratosphere')
ax1.plot(xran,[15,15],'k--')
ax1.text(400, 5, 'Troposphere')
title('Neutral gas')

ax2 = fig.add_axes([0.56, 0.1, 0.4, 0.85])
p21, = ax2.plot(nelNight, alt,'b', label='Night')
p22, = ax2.plot(nelDay, alt,'r', label='Day')
ax2.legend(loc=4,prop={'size':13})
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.yaxis.set_ticks_position('both')
ax2.yaxis.set_major_formatter(NullFormatter())
ax2.set_xlabel('Plasma density [m$^{-3}$]')
xran = [9e8,2e12]
ax2.set_xlim(xran)
ax2.set_ylim([1,2e3])
ax2.text(4e9, 300., 'F region (EUV)')
ax2.plot(xran,[180,180],'k--')
ax2.text(4e9, 120., 'E region (Soft X-rays)')
ax2.plot(xran,[85,85],'k--')
ax2.text(4e9, 65., r'D region (Ly $\alpha$, Hard X-rays)')
ax2.annotate('', 
            xy=(5e9, 1.9e3), xycoords='data',
            xytext=(5e9, 1.3e3), textcoords='data',
            arrowprops=dict(arrowstyle="simple",
                            color="0.5",
                            connectionstyle="arc3"),
            )
ax2.text(2e9, 1.1e3, 'Plasmasphere')
title('Ionized gas')
savefig('ionos.pdf')
```
