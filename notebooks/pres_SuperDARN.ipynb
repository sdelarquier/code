```python
import sys
sys.path.append('/davitpy')
import pydarn
from datetime import datetime
```

```python
rads = pydarn.radar.network()
dtnow = datetime.utcnow()
midlat = []
highlat = []
polar = []
for rad in rads.info:
    tcod = rad.getSiteByDate(dtnow)
    if tcod and rad.status == 1 and rad.stTime <= dtnow <= rad.edTime and tcod.geolat >= 0:
        if tcod.geolat >= 70.: polar.append(rad.code[0])
        elif 50. <= tcod.geolat < 70.: highlat.append(rad.code[0])
        else: midlat.append(rad.code[0])
```

```python
figure(figsize=(8,8))
myMap = pydarn.plot.map(boundingLat=30)
pydarn.plot.overlayRadar(myMap, all=True, hemi='north')
pydarn.plot.overlayFov(myMap, codes=midlat, hemi='north', maxGate=75, lineWidth=.5, fovColor='r')
pydarn.plot.overlayFov(myMap, codes=highlat, hemi='north', maxGate=75, lineWidth=.5, fovColor='b')
pydarn.plot.overlayFov(myMap, codes=polar, hemi='north', maxGate=75, lineWidth=.5, fovColor='g')
```
