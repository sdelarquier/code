```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = mpimg.imread('/home/sebastien/Desktop/img_map2.png')
```

####Separate R, G, and B channels

```python
imgR = img[:,:,0]/sum(img, axis=2, dtype='f')*256
imgG = img[:,:,1]/sum(img, axis=2, dtype='f')*256
imgB = img[:,:,2]/sum(img, axis=2, dtype='f')*256
```

```python
figure(figsize=(14,8))
subplot(221)
imgplot = plt.imshow(img)
title('orginial')
subplot(222)
imgplotR = plt.imshow(imgR, vmin=0, vmax=255, cmap='binary')
title('red')
subplot(223)
imgplotG = plt.imshow(imgG, vmin=0, vmax=255, cmap='binary')
title('green')
subplot(224)
imgplotB = plt.imshow(imgB, vmin=0, vmax=255, cmap='binary')
title('blue')
savefig('/home/sebastien/Desktop/img_.png')
```

####Build histogram for each channel

```python
figure(figsize=(14,4))
freqMax = 1e4
subplot(131)
imghistR = plt.hist(imgR.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('red')
subplot(132)
imghistG = plt.hist(imgG.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('green')
subplot(133)
imghistB = plt.hist(imgB.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('blue')
```

####Stretch histogram

```python
# Linear stretch transform between 50 and 125
hmod = lambda x: 3.4*(x.copy() - 50)
```

```python
imgRmod = hmod(imgR)
imgGmod = hmod(imgG)
imgBmod = hmod(imgB)
imgmod = img.copy()
imgmod[:,:,0] = imgRmod*sum(img, axis=2, dtype='f')/256
imgmod[:,:,1] = imgGmod*sum(img, axis=2, dtype='f')/256
imgmod[:,:,2] = imgBmod*sum(img, axis=2, dtype='f')/256
```

```python
figure(figsize=(14,4))
freqMax = 3e3
subplot(131)
imghistR = plt.hist(imgRmod.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('red')
subplot(132)
imghistG = plt.hist(imgGmod.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('green')
subplot(133)
imghistB = plt.hist(imgBmod.flatten(), 256, range=(0,255), fc='k', ec='k')
ylim([0,freqMax])
title('blue')
```

```python
figure(figsize=(14,8))
subplot(221)
imgplot = plt.imshow(imgmod)
title('orginial stretched')
subplot(222)
imgplotR = plt.imshow(imgRmod, vmin=0, vmax=255, cmap='binary')
title('red')
subplot(223)
imgplotG = plt.imshow(imgGmod, vmin=0, vmax=255, cmap='binary')
title('green')
subplot(224)
imgplotB = plt.imshow(imgBmod, vmin=0, vmax=255, cmap='binary')
title('blue')
savefig('/home/sebastien/Desktop/img_hist.png')
```

####Plot spectral distributions

```python
figure(figsize=(14,4))
subplot(131)
scatter(imgRmod.flatten(), imgGmod.flatten(), marker='.', s=1)
xlim([0,255])
ylim([0,255])
xlabel('red')
ylabel('green')
subplot(132)
scatter(imgGmod.flatten(), imgBmod.flatten(), marker='.', s=1)
xlim([0,255])
ylim([0,255])
xlabel('green')
ylabel('blue')
subplot(133)
scatter(imgBmod.flatten(), imgRmod.flatten(), marker='.', s=1)
xlim([0,255])
ylim([0,255])
xlabel('blue')
ylabel('red')
```

```python
# Calculate covariance matrix
Co = np.cov([imgRmod.flatten(), imgGmod.flatten(), imgBmod.flatten()])
```

```python
# Calculate eigenvalues and eigenvectors
eigenval, eigenvec = np.linalg.eig(Co)
```

```python
# Build with eigenvectors ordered by decreasing eigenvalues
A = eigenvec[np.argsort(eigenval)[::-1],:]
```

```python
imgPCA = zeros(shape(img))
for i in xrange(shape(img)[0]):
    for j in xrange(shape(img)[1]):
        imgPCA[i,j,:] = A.dot(imgmod[i,j,:])
imgPCA = imgPCA.astype('float32')
```

```python
imgplot = plt.imshow(imgPCA)
```
