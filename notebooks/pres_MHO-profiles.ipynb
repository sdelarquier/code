```python
import sys
sys.path.append('/davitpy')
from gme.isr import *
from models import aacgm
import datetime as dt
```

```python
expDate = dt.datetime(2006, 2, 22, 22)
dataPath = '/home/sebastien/Documents/code/mho/data/'
user_fullname = 'Sebastien de Larquier'
user_email = 'sdelarquier@vt.edu'
user_affiliation = 'Virginia Tech'
data = mhoData( expDate, dataPath=dataPath, 
            user_fullname=user_fullname, 
            user_email=user_email, 
            user_affiliation=user_affiliation )
```

```python
look = data.look()
print look
```

```python
i = 2

tlim = [date2num(data.time[0].replace(second=0, minute=0)), 
        date2num(data.time[-1].replace(second=0, minute=0, hour=data.time[-1].hour+1))]
alim = [100,600]


inds = (np.around(data.elev[:,0],1) == look[i][0]) & (np.around(data.azim[:,0],1) == look[i][1])
vinds = np.isfinite(data.gdalt[:,inds][:,0])

figure(figsize=(10,8))
subplots_adjust(hspace=0.12)

ax = subplot(211)
rcParams.update({'font.size': 14})
x, y = meshgrid(date2num(data.time[inds]), data.range[vinds])
im = ax.pcolormesh(x, data.gdalt[:,inds][vinds,:], ma.masked_invalid( data.nel[:,inds][vinds,:] ), vmin=9, vmax=12)
ylim(alim)
xlim(tlim)
xticks = HourLocator(interval=1)
xticksfmt = DateFormatter('%H:%M')
ax.xaxis.set_major_locator(xticks)
ax.xaxis.set_major_formatter(xticksfmt)
xticksminor = MinuteLocator(interval=10)
ax.xaxis.set_minor_locator(xticksminor)
minorLocator = MultipleLocator(100)
ax.yaxis.set_minor_locator(minorLocator)
ylabel('Altitude [km]')
cb = colorbar(im, pad=0.1)
cb.set_label(r'Electron density [$\log(m^{-3})$]')
title(r'El: {:.1f}$^\circ$, Az: {:.1f}$^\circ$E'.format(look[i][0], look[i][1]))
ax.set_xticklabels(ax.get_xticklabels(), visible=False)
ax.grid()
axr = ax.twinx()
axr.set_ylabel( 'Magn. Latitude' )
yticks = ax.get_yticks()
mlat, _, _ = aacgm.aacgmConvArr(list(data.lat[:,inds][vinds,:][:,0]), 
                                list(data.lon[:,inds][vinds,:][:,0]), 
                                list(data.gdalt[:,inds][vinds,:][:,0]), 0)
yticks2 = np.interp(yticks, data.gdalt[:,inds][vinds,:][:,0], mlat, left=0, right=0)
axr.set_yticks(yticks)
axr.set_ylim([yticks[0], yticks[-1]])
axr.set_yticklabels(['{:.1f}'.format(yt) if yt!=0 else '' for yt in yticks2])

ax = subplot(212)
rcParams.update({'font.size': 14})
im = ax.pcolormesh(x, data.gdalt[:,inds][vinds,:], ma.masked_invalid( data.te[:,inds][vinds,:] ), vmin=400, vmax=4000)
ylim(alim)
xlim(tlim)
xticks = HourLocator(interval=1)
xticksfmt = DateFormatter('%H:%M')
ax.xaxis.set_major_locator(xticks)
ax.xaxis.set_major_formatter(xticksfmt)
xticksminor = MinuteLocator(interval=10)
ax.xaxis.set_minor_locator(xticksminor)
minorLocator = MultipleLocator(100)
ax.yaxis.set_minor_locator(minorLocator)
ax.tick_params(axis='x', pad=12)
xlabel('Time [UT]')
ylabel('Altitude [km]')
cb = colorbar(im, pad=0.1)
cb.set_label(r'Electron temperature [$K$]')
ax.grid()
axr = ax.twinx()
axr.set_ylabel( 'Magn. Latitude' )
yticks = ax.get_yticks()
mlat, _, _ = aacgm.aacgmConvArr(list(data.lat[:,inds][vinds,:][:,0]), 
                                list(data.lon[:,inds][vinds,:][:,0]), 
                                list(data.gdalt[:,inds][vinds,:][:,0]), 0)
yticks2 = np.interp(yticks, data.gdalt[:,inds][vinds,:][:,0], mlat, left=0, right=0)
axr.set_yticks(yticks)
axr.set_ylim([yticks[0], yticks[-1]])
axr.set_yticklabels(['{:.1f}'.format(yt) if yt!=0 else '' for yt in yticks2])

savefig('/home/sebastien/Dropbox/mho.{:%Y%m%d}.{:02.0f}.{:03.0f}.png'.format(expDate, look[i][0], look[i][1]))
```
