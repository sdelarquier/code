```python
# Create a mask for required pointing and ltitude conditions
def selectPointing(mho, azim=None, elev=None, param='edens'):
    cond_prm = (mho[param] > 0)
    if elev:
        cond_el = logical_and(mho['elev'][:,0] > elev-.5, mho['elev'][:,0] < elev+.5)
        cond_el = tile( cond_el, (shape(mho[param])[0],1) )
    else: cond_el = cond_prm
    if azim:
        cond_az = logical_and(mho['azim'][:,0] > azim-.5, mho['azim'][:,0] < azim+.5)
        cond_az = tile( cond_az, (shape(mho[param])[0],1) )
    else: cond_az = cond_prm
    return cond_el*cond_az*cond_prm

# Plot data (2D: altitude-time)
def mhoPlotAT(mho, azim=None, elev=None, alti=[0., 500.], time=None, param='edens', scale=None):
    conds = selectPointing(mho, azim=azim, elev=elev, param=param)
    if time is None:
        stime = expDate.replace(hour=19)
        etime = expDate.replace(hour=13) + timedelta(days=1)
        time = date2num([stime,etime])
    # Set scale
    if scale is None: scale = [mho[param][conds].min(), mho[param][conds].max()]
    # Get axis
    ax = gca()
    
    # Create a polygon collection
    height = 10
    patches = []
    values = []
    for j in xrange(shape(mho['alti'])[1]-1):
        if mho['time'][j] < time[0]: continue
        if mho['time'][j] > time[1]: continue
        width = mho['time'][j+1] - mho['time'][j]
        for i in xrange(shape(mho['alti'])[0]):
            if not conds[i,j]: continue
            rect = mp.patches.Rectangle( xy=(mho['time'][j], mho['alti'][i,j]), width=width, height=height )
            patches.append( rect )
            values.append( mho[param][i,j]  )
    cax = mp.collections.PatchCollection(patches, cmap=mp.cm.jet, linewidths=0, norm=mp.colors.Normalize(vmin=scale[0], vmax=scale[1]))
    cax.set_array(array(values))
    im = ax.add_collection(cax)
    
    x, y = meshgrid(mho['time'], mho['range'])
    xticks = mp.dates.HourLocator(interval=2)
    xticksfmt = mp.dates.DateFormatter('%H:%M')
    ax.xaxis.set_major_locator(xticks)
    ax.xaxis.set_major_formatter(xticksfmt)
    ax.set_xlabel('Time [UT]')
    ax.set_ylabel('Altitude [km]')
    if alti is not None: ylim(alti)
    if time is not None: xlim(time)
    grid()
    # Color bar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", "3%", pad="3%")
    cbar = colorbar(im, cax=cax)
    cbar.set_label(mlhRead.mhoCipher[param])
```

```python
def plotViews(expDate, misa=None, azim=None):
    mho = mlhRead.mhoRead(expDate)
    if not mho: return
    
    dw = 111e3*55
    myMap = plu.mapObj(lat_0=50., lon_0=-100., width=dw, height=dw)
    
    # MHO location
    if not azim: 
        inds = where(mho['scntyp'] == 2)
        azim = [mho['azim'][inds,:].min(), mho['azim'][inds,:].max()]
    if not misa: 
        inds = where( (mho['scntyp'] == 1) & (mho['elev'][:,0] < 85.) )
        misa = mean(mho['azim'][inds])
    mlhRead.isrFov(myMap, 'MHO', mho['pos'][0:2], 12, azim=azim, misa=misa)
    
    fovCol = (.5,.6,.6)
    fovAlpha = .2
    overlayFov(myMap, codes=['bks'], hemi='north', maxGate=75, dateTime=expDate, 
            fovColor=fovCol, lineColor='.2', lineWidth=.5, fovAlpha=fovAlpha)
    overlayRadar(myMap, codes=['bks'], fontSize=10)
    
    return myMap
```

### Plot BKS RTIs with MHO data

```python
def plotRTI(data, info, expDate, beam=7, channel=0, param='power', scale=[0, 30], time=[2100, 1300], ngates=70, noX=False, noTitle=False, fontsize=12):
    beaminds = ( data.beam == beam )
    chaninds = ( data.channel == channel )
    datinds = np.logical_and(beaminds, chaninds)
    tdelta = timedelta(days=1) if time[1] <= time[0] else timedelta(days=0)
    stime = mp.dates.date2num(expDate.replace(hour=time[0]/100, minute=(time[0]-time[0]/100*100)))
    ftime = mp.dates.date2num(expDate.replace(hour=time[1]/100, minute=(time[1]-time[1]/100*100))+tdelta)
    xtimes = mp.dates.julian2num(data.juls)
    timinds = np.logical_and( xtimes >= stime, xtimes <= ftime)
    inds = np.logical_and( datinds, timinds )
    
    # Plotting
    ax = gca()
    x = xtimes[inds]
    y = np.arange(data.lagfr[0]*.15, data.lagfr[0]*.15+ngates*data.smsep[0]*.15, data.smsep[0]*.15)
    X, Y = meshgrid(x, y)
    Z = data[param][:,inds]
    im = ax.pcolormesh(X, Y, np.ma.masked_where(Z == 1e4, Z), vmin=scale[0], vmax=scale[1])
    
    # Formatting
    xticks = mp.dates.HourLocator(interval=2)
    xticksfmt = mp.dates.DateFormatter('%H:%M')
    ax.xaxis.set_major_locator(xticks)
    ax.xaxis.set_major_formatter(xticksfmt)
    xlim([stime, ftime])
    ylim([0, y[-1]])
    if noX: plt.setp(ax.get_xticklabels(), visible=False)
    if not noX: ax.set_xlabel('Time [UT]')
    ax.set_ylabel('Range [km]')
    ax.grid()
    ax.text(.5, .9, 'Beam {:02d}'.format(beam), transform=ax.transAxes, zorder=3, weight='heavy')
    
    # Day/Night
    fov = radar.fov(site=radar.network().getRadarByCode(info.code).getSiteByDate(expDate))
    lats = fov.latCenter[beam,:]
    lons = fov.lonCenter[beam,:]
    triseP, tsetP = calcSun.calcSunRiseSetUTC(calcSun.getJD(expDate-timedelta(days=1)), lats, lons)
    trise0, tset0 = calcSun.calcSunRiseSetUTC(calcSun.getJD(expDate), lats, lons)
    triseF, tsetF = calcSun.calcSunRiseSetUTC(calcSun.getJD(expDate+timedelta(days=1)), lats, lons)
    tsetP = mp.dates.date2num( [expDate.replace(hour=0, minute=0, second=0) - timedelta(days=1) + timedelta(minutes=tset) for tset in tsetP] )
    trise0 = mp.dates.date2num( [expDate.replace(hour=0, minute=0, second=0) + timedelta(minutes=trise) for trise in trise0 ] )
    tset0 = mp.dates.date2num( [expDate.replace(hour=0, minute=0, second=0) + timedelta(minutes=tset) for tset in tset0 ] )
    triseF = mp.dates.date2num( [expDate.replace(hour=0, minute=0, second=0) + timedelta(days=1) + timedelta(minutes=trise) for trise in triseF ] )
    term1 = concatenate([tsetP, trise0[-1::-1]])
    term2 = concatenate([tset0, triseF[-1::-1]])
    crang = concatenate([fov.slantRCenter[beam,:], fov.slantRCenter[beam,-1::-1]])
    contour = transpose( vstack((term1, crang)) )
    patch1 = mp.patches.Polygon( contour, color='0.7', alpha=.5, zorder=0)
    ax.add_patch(patch1)
    contour = transpose( vstack((term2, crang)) )
    patch2 = mp.patches.Polygon( contour, color='0.7', alpha=.5, zorder=0)
    ax.add_patch(patch2)
    plot(tsetP, fov.slantRCenter[beam,:], 'k--')
    plot(trise0, fov.slantRCenter[beam,:], 'k--')
    plot(tset0, fov.slantRCenter[beam,:], 'k--')
    plot(triseF, fov.slantRCenter[beam,:], 'k--')
    
    # Colorbar
    if not noTitle: title('{} - {}'.format(expDate.strftime('%Y/%b/%d'), info.name))
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", "3%", pad="3%")
    cbar = colorbar(im, cax=cax)
    cbar.set_label(param)
    mp.rcParams.update({'font.size': fontsize})
```
