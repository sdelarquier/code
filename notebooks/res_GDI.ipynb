```python
def pcolormeshISR(data, param='nel', title=None, **kwargs):
    if not gcf(): figure(figsize=(7,4))
    X, Y = meshgrid(data['time'], data['range'])
    #datplot = signal.medfilt(data[param])
    datplot = data[param]
    pcolormesh(X, data['gdalt'], ma.masked_invalid(datplot), **kwargs)
    xticks = dates.HourLocator(interval=2)
    xticksfmt = dates.DateFormatter('%H:%M')
    gca().xaxis.set_major_locator(xticks)
    gca().xaxis.set_major_formatter(xticksfmt)
    expDate = dates.num2date(data['time'][0])
    print expDate
    xlim([expDate.replace(hour=22, minute=0),expDate.replace(hour=12, minute=0)+timedelta(days=1)])
    ylim([100.,500.])
    xlabel('time [UT]')
    ylabel('Alt. [km]')
    colorbar()
    if title: gca().title(title)
```

```python
figure(figsize=(20,4))
pcolormeshISR(dat1, vmin=8, vmax=12)
```
