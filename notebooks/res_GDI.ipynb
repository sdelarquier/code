```python
def pcolormeshISR(data, inds, param='edens', **kwargs):
    if not gcf(): figure(figsize=(7,4))
    X, Y = meshgrid(data['time'][inds], data['range'])
    pcolormesh(X, data['alti'][:,inds], ma.masked_invalid(signal.medfilt(data[param][:,inds])), **kwargs)
    xticks = dates.HourLocator(interval=2)
    xticksfmt = dates.DateFormatter('%H:%M')
    gca().xaxis.set_major_locator(xticks)
    gca().xaxis.set_major_formatter(xticksfmt)
    expDate = dates.num2date(data['time'][inds][0])
    print expDate
    xlim([expDate.replace(hour=22, minute=0),expDate.replace(hour=12, minute=0)+timedelta(days=1)])
    ylim([100.,500.])
    xlabel('time [UT]')
    ylabel('Alt. [km]')
    colorbar()
    title(mhoCipher[param])
```

```python
figure(figsize=(20,4))
subplot(131)
pcolormeshISR(dat1, indsUp, vmin=8, vmax=12)
subplot(132)
pcolormeshISR(dat1, indsSc, vmin=8, vmax=12)
subplot(133)
pcolormeshISR(dat1, indsFx, vmin=8, vmax=12)
```
