```python
# Experiment date
expDate = dt.datetime(2006,2,22,21)
# Read data from hdf5 files (the default hdf5 from Madrigal enhanced with corrected electron densities from isprint)
data = mhoRead(expDate, fileExt='g.003')
#data = mhoRead(expDate, fileExt='a.000')
# Altitude
alt = 300.
```

```python
for i in range(len(data['time'])):
    if data['elev'][i,0] > 80.: continue
    print num2date( data['time'][i] ), '\t', data['elev'][i,0]
    for j in range(len(data['range'])):
        if isfinite(data['te'][j,i]) and isfinite(data['nel'][j,i]):
            print '\t{:8.2f}\t{:8.2f} ({:8.2f})\t{:8.2f}\t{:8.2f}'.format( data['range'][j], 
                                                                           data['alti'][j,i], 
                                                                           data['gdalt'][j,i], 
                                                                           10.**(data['nel'][j,i]-10), 
                                                                           data['te'][j,i] )
```

```python
alts = [alt-40, alt-20, alt]
dalt = 10.
# Select all 4 elevations (we will bin Ne and Te by time and elevation)
elevs = sort(unique(data['elev'].round()))
# Init lists (elevation binned lists)
azim = []
mlat, mlon, glat, glon = [], [], [], []
tt, ne, te = [], [], []
# Loop through each elevation
for el in elevs:
    # Empty temp lists (time series lists)
    elazim = []
    eltt, elne, elte = [], [], []
    elmlat, elmlon, elglat, elglon = [], [], [], []
    for k in range(len(alts)):
        alt = alts[k]
        kelazim = []
        keltt, kelne, kelte = [], [], []
        kelmlat, kelmlon, kelglat, kelglon = [], [], [], []
        # Loop through time
        for i, time in enumerate(data['time'][:-1]):
            # Only look at data at the current elevation
            if not round(data['elev'][i,0]) == el: continue
            # Select data close to 300 km altitude
            indsAlt = np.isfinite(data['nel'][:,i]) & np.isfinite(data['te'][:,i])
            ialt = np.argmin( abs( data['gdalt'][indsAlt,i] - alt ) )
            # Append to time series lists (mean values of data points found between 290-310 km altitude)
            #tkelne = 10.**( data['nel'][indsAlt,i][ialt] - 10. )
            #tkelte = ( data['te'][indsAlt,i][ialt] )
            tkelne = 10.**( np.interp( alt, data['gdalt'][indsAlt,i], data['nel'][indsAlt,i] ) - 10. )
            tkelte = np.interp(alt, data['gdalt'][indsAlt,i], data['te'][indsAlt,i])
            #print num2date(time).replace(tzinfo=None), el, alts[k], data['gdalt'][indsAlt,i][ialt], tkelne, tkelte
            if not isfinite(tkelne) or not isfinite(tkelte): continue
            kelne.append( tkelne )
            kelte.append( tkelte )
            # Calculate geomagnetic latitude
            tkelmlat, tkelmlon, _ = aacgm.aacgmConv( mean(data['lat'][indsAlt,i]), 
                                                     mean(data['lon'][indsAlt,i]), 
                                                     mean(data['alti'][indsAlt,i]), 0)
            kelmlat.append( tkelmlat )
            kelmlon.append( tkelmlon )
            kelglat.append( mean(data['lat'][indsAlt,i]) )
            kelglon.append( mean(data['lon'][indsAlt,i]) )
           
            keltt.append( data['time'][i] )
            kelazim.append( data['azim'][i] )
        #print '---------'
        eltt = keltt
        elazim = kelazim

        elne.append( kelne )
        elte.append( kelte )
        elmlat.append( mean(kelmlat) )
        elmlon.append( mean(kelmlon) )
        elglat.append( mean(kelglat) )
        elglon.append( mean(kelglon) )
    # Append time series to elev binned lists
    azim.append(mean(elazim))
    tt.append( eltt )
    te.append( elte )
    ne.append( elne )
    mlat.append( elmlat )
    mlon.append( elmlon )
    glat.append( elglat )
    glon.append( elglon )
mlat = array(mlat)
glat = array(glat)
glon = array(glon)
```

```python
# Best common time sample:
nsamples = min(len(tt[i]) for i in range(len(tt)))
minTime = min(min(tt[i]) for i in range(len(tt)))
maxTime = max(max(tt[i]) for i in range(len(tt)))
time_sample = linspace(minTime, maxTime, nsamples)
# Linear 1D interpolation of ne and te to have the same sampling rate at all elevation
neRS, teRS = [], []
for i in range(len(elevs)):
    tneRS, tteRS = [], []
    for k in range(len(alts)):
        tneRS.append( np.interp(time_sample, tt[i], ne[i][k]) )
        tteRS.append( np.interp(time_sample, tt[i], te[i][k]) )
    neRS.append(tneRS)
    teRS.append(tteRS)
neRS = array(neRS)
teRS = array(teRS)
```

```python
# Ray did not use the lowest elevation in his gradients (why?)
# Elevation angles are sorted, so we start at 1 (skip 0)
start = 1
# Initialize arrays
neScale = zeros((len(time_sample),3))
teScale = zeros((len(time_sample),3))
neGrad = zeros((len(time_sample),3))
teGrad = zeros((len(time_sample),3))
alt = alts[-1]
ialt = alts.index(alt)
# Loop through time
for i in range(len(time_sample)):
    # density gradients (horizontal)
    x = mlat[start:,ialt]
    y = neRS[start:,ialt,i]
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0] # linear least square fit
    neScale[i,0] = ( mean(y)/m )*122. # convert the latitude scale length to km (crude)
    neGrad[i,0] = m/122.
    # temperature gradient (horizontal)
    y = teRS[start:,ialt,i]
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0] # linear least square fit
    teScale[i,0] = ( mean(y)/m )*122. # convert the latitude scale length to km (crude)
    teGrad[i,0] = m/122.
    # density gradients (vertical)
    x = alts
    y = neRS[-1,:,i] #mean(neRS[:,:,i], axis=0)
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0] # linear least square fit
    neScale[i,1] = ( mean(y)/m ) # convert the latitude scale length to km (crude)
    neGrad[i,1] = m
    # temperature gradient (vertical)
    y = teRS[-1,:,i] #mean(teRS[:,:,i], axis=0)
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0] # linear least square fit
    teScale[i,1] = ( mean(y)/m ) # convert the latitude scale length to km (crude)
    teGrad[i,1] = m
```

```python
# Calculate dip and declination of B field within region
# INPUTS
itype = 1 # Geodetic coordinates
date = utils.dateToDecYear(expDate) # decimal year
mglat = mean(glat[:,-1])
mglon = mean(glon[:,-1])
xlti, xltf, xltd = mglat,mglat,1. # latitude start, stop, step
xlni, xlnf, xlnd = mglon,mglon,1. # longitude start, stop, step
ifl = 0 # Main field
# Call fortran subroutine
lat,lon,d,s,h,x,y,z,f = igrf.igrf11(itype,date,alt,ifl,xlti,xltf,xltd,xlni,xlnf,xlnd)

# Conpute gradients perpendicular to B
neGrad[:,2] = neGrad[:,0] * np.sin( np.radians(s[0]) ) + neGrad[:,1] * np.cos( np.radians(s[0]) )
neScale[:,2] = mean(mean(neRS, axis=0), axis=0) / neGrad[:,2]
teGrad[:,2] = teGrad[:,0] * np.sin( np.radians(s[0]) ) + teGrad[:,1] * np.cos( np.radians(s[0]) )
teScale[:,2] = mean(mean(teRS, axis=0), axis=0) / teGrad[:,2]

dip = s[0]
print lat,lon%360.,d,s,h,x,y,z,f
```
