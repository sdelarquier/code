```python
def plotIri(dat):
    X, Y = meshgrid(linspace(0, 2500, 250), linspace(60, 560, 50))
    pcolormesh( X, Y, transpose(dat.rt_data['edens'][0][:,:,0,0]), vmin=1e8, vmax=2e11 )
    mask = (dat.rt_data['power'][0][:,0,0] == 10000)
    scatter( ma.masked_array(dat.rt_data['grange'][0][:,0,0], mask=mask), ma.masked_array(dat.rt_data['altitude'][0][:,0,0], mask=mask), c='k' )
    xlim([0,2500])
    ylim([0,560])
    power = dat.rt_data['power'][0][(dat.rt_data['power'][0][:,0,0] != 10000),0,0]
    altitude = dat.rt_data['altitude'][0][(dat.rt_data['power'][0][:,0,0] != 10000),0,0]
    grange = dat.rt_data['grange'][0][(dat.rt_data['power'][0][:,0,0] != 10000),0,0]
    srange = arange(180., 180.+71*45., 45. )[(dat.rt_data['power'][0][:,0,0] != 10000)]
    return srange[power.argmax()], grange[power.argmax()], altitude[power.argmax()]
```
