```python
import datetime as dt
import numpy as np
import matplotlib as mp
```

```python
# Create a date vector
dates = [dt.datetime(2010, 1, 1)]
stop = dt.datetime(2011, 1, 1)
step = dt.timedelta(days=1)
while dates[-1]+step < stop:
    dates.append(dates[-1]+step)
```

```python
# Generate random data
data = randn(len(dates))
```

```python
fig = figure( figsize=(10,5) )
ax = fig.add_subplot(111)
rcParams.update({'font.size': 12})
# X axis
xlabel('Day-of-Month')
xlim( (0, 31) )
# Y axis
ylabel('Month')
ylim( (0, 12) )
ax.set_yticks( arange(0,12,1) )
month = [dt.date(1970, m, 1).strftime('%b') for m in arange(1,13,1)]
ax.set_yticklabels(month, va='bottom')

# Plotting the data
# Create a polygon collection
width = 1
height = 1
patches = []
for date in dates:
    rect = mp.patches.Rectangle( xy=(date.day-1, date.month-1), width=width, height=height )
    patches.append( rect )
p = mp.collections.PatchCollection(patches, cmap=mp.cm.jet, linewidths=2, edgecolors='w')
p.set_array(data)
ax.add_collection(p)
colorbar(p)
```
