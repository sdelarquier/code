```python
def mhoFov(myMap, elev, dist, azim=[-90.,15.]):
    # MHO location
    mhpos = [42.6, 288.5]
    x0, y0 = myMap(mhpos[1], mhpos[0])
    myMap.scatter(x0, y0, s=20, zorder=5, c='k')
    text(x0*1.04, y0*0.96, 'MHO')
    
    # MHO fov
    nazims = 100
    azims = linspace(azim[0], azim[1], nazims)
    mhfov = zeros((nazims, 3))
    Rav = 6370.
    for iaz, taz in enumerate(azims):
        dd = gp.calcDistPnt(mhpos[0], mhpos[1], 0., dist=dist, el=elev, az=taz)
        mhfov[iaz,0] = dd['distLat']
        mhfov[iaz,1] = dd['distLon']
        mhfov[iaz,2] = dd['distAlt']
    x, y = myMap(mhfov[:,1], mhfov[:,0])
    myMap.plot(x, y, 'g')
    myMap.plot([x0,x[0]], [y0,y[0]], 'g', zorder=3)
    myMap.plot([x0,x[-1]], [y0,y[-1]], 'g', zorder=3)
    
    return mhfov
```

```python
radCode = 'wal'
sTime = dt.datetime(2006,2,23,0,0)
eTime = dt.datetime(2006,2,23,5,0)
myPtrBks = pydarn.sdio.radDataOpen(sTime,radCode,eTime=eTime,fileType='fitacf',filtered=True)
#myPtrWal = pydarn.sdio.radDataOpen(sTime,'wal',eTime=eTime,channel=channel,bmnum=bmnum,cp=cp,fileType=fileType,filtered=True, src=src)
```

```python
from pydarn.radar import *
bks = radar(code=radCode)
siteBks = bks.getSiteByDate(sTime)
statsBks = zeros((siteBks.maxbeam, siteBks.maxgate))
```

```python
myBeam = pydarn.sdio.radDataReadRec(myPtrBks)
beams = {}
while myBeam:
    if myBeam.bmnum not in beams: beams[myBeam.bmnum] = 1
    else: beams[myBeam.bmnum] += 1
    if myBeam.fit.slist:
        for i,gate in enumerate(myBeam.fit.slist):
            if myBeam.fit.p_l[i] <= 5: continue
            statsBks[myBeam.bmnum, gate] += myBeam.fit.p_l[i]
    myBeam = pydarn.sdio.radDataReadRec(myPtrBks)
for b in beams.keys():
    statsBks[b, :] = statsBks[b, :]/beams[b]
statsBks = statsBks/statsBks.max()
```

```python
fovBks = fov(site=siteBks, ngates=70, coords='mag')
```

```python
def overlayData(myMap, data, lats, lons):
    scale = [data.min(),data.max()]
    # Get axis
    ax = gca()
    
    patches = []
    values = []
    for i in xrange(len(lats[:,0])-1):
        for j in xrange(len(lats[0,:])-1):
            if data[i,j] == 0: continue
            contourLat = [lats[i,j], lats[i+1,j], lats[i+1,j+1], lats[i,j+1]]
            contourLon = [lons[i,j], lons[i+1,j], lons[i+1,j+1], lons[i,j+1]]
            x, y = myMap(contourLon, contourLat, coords='mag')
            rect = mp.patches.Polygon( xy=transpose( vstack((x,y)) ) )
            patches.append( rect )
            values.append( data[i,j]  )
            
    cax = mp.collections.PatchCollection(patches, cmap=mp.cm.jet, linewidths=0, edgecolors='none', 
                    norm=mp.colors.Normalize(vmin=scale[0], vmax=scale[1]), zorder=2, alpha=.6)
    cax.set_array(array(values))
    im = ax.add_collection(cax)

    # More stuff
    rcParams.update({'font.size': 12})
    Title = '{} to {} UT'.format(sTime.strftime('%Y / %b / %d - %H:%M'), eTime.strftime('%H:%M'))
    text(myMap2.llcrnrx, myMap2.urcrnry, 
        Title, rotation=0, va='bottom', ha='left')
    
    # Color bar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", "3%", pad="3%")
    cbar = colorbar(im, cax=cax)
    cbar.set_label('Scatter distribution')
```
