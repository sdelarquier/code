### Generate a text file with footpoints

```python
sdate = datetime(2012,8,30)
edate = datetime(2013,12,31)
date = sdate
dates = [date]
while date < edate:
    date = dates[-1]+timedelta(days=1)
    dates.append(date)
```

```python
def addDay(date, output):
    traceA = ts.tsygTrace(filename='../rbsp/data/trace.{}.A.dat'.format(date.strftime('%Y%m%d')))
    Adtime = traceA.datetime
    AlatN = traceA.latNH
    AlonN = traceA.lonNH
    AlatS = traceA.latSH
    AlonS = traceA.lonSH
    traceA = 0
    traceB = ts.tsygTrace(filename='../rbsp/data/trace.{}.B.dat'.format(date.strftime('%Y%m%d')))
    Bdtime = traceB.datetime
    BlatN = traceB.latNH
    BlonN = traceB.lonNH
    BlatS = traceB.latSH
    BlonS = traceB.lonSH
    traceB = 0
    for i in range(len(Adtime)):
        year   = Adtime[i].year
        month  = Adtime[i].month
        day    = Adtime[i].day
        hour   = Adtime[i].hour
        minute = Adtime[i].minute
        sat = 'A'
        latN = AlatN[i]
        lonN = AlonN[i]
        latS = AlatS[i]
        lonS = AlonS[i]
        output.append('{:4d} {:2d} {:2d} {:2d} {:2d} {} {:6.2f} {:6.2f} {:6.2f} {:6.2f}'.format(year, month, day, hour, minute, sat, latN, lonN, latS, lonS))
        
        year   = Bdtime[i].year
        month  = Bdtime[i].month
        day    = Bdtime[i].day
        hour   = Bdtime[i].hour
        minute = Bdtime[i].minute
        sat = 'B'
        latN = BlatN[i]
        lonN = BlonN[i]
        latS = BlatS[i]
        lonS = BlonS[i]
        output.append('{:4d} {:2d} {:2d} {:2d} {:2d} {} {:6.2f} {:6.2f} {:6.2f} {:6.2f}'.format(year, month, day, hour, minute, sat, latN, lonN, latS, lonS))
```

```python
output = []
jobs = []
for date in dates:
    addDay(date, output)
```

### Generate a text file with DB-stored apogee footpoints

```python
import sqlite3 as lite
import sys

con = lite.connect('/var/www/assets/db/rbsp.db', detect_types=lite.PARSE_DECLTYPES)
with con:
    con.row_factory = lite.Row
    cur = con.cursor()

    # Select data from db
    cur.execute("SELECT sat,time,lat,lon FROM rbsp_apogee ORDER BY time")
    data = cur.fetchall()
    
    npts = len(data)
    output = []
    for i in range(0,npts,2):
        year = data[i]['time'][0:4]
        month = data[i]['time'][5:7]
        day = data[i]['time'][8:10]
        hour = data[i]['time'][11:13]
        minute = data[i]['time'][14:16]
        sat = data[i]['sat']
        if data[i]['lat'] > 0.:
            latN = data[i]['lat']
            lonN = data[i]['lon']
            latS = data[i+1]['lat']
            lonS = data[i+1]['lon']
        else:
            latN = data[i+1]['lat']
            lonN = data[i+1]['lon']
            latS = data[i]['lat']
            lonS = data[i]['lon']
        output.append('{} {} {} {} {} {} {:6.2f} {:6.2f} {:6.2f} {:6.2f}'.format(year, month, day, hour, minute, sat, latN, lonN, latS, lonS))
```
