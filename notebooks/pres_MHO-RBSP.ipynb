```python
from datetime import datetime as dtmod
from mpl_toolkits.basemap import Basemap
import tsyganenko as ts
import sys
sys.path.append('/davitpy')
from utils import geoPack as gp
from pydarn.plot import overlayFov
from pydarn.plot import overlayRadar
```

```python
# Select a date
date = dtmod(2012,11,28)
# Load RBSP position
traceA = ts.tsygTrace(filename='../rbsp/data/trace.{}.A.dat'.format(date.strftime('%Y%m%d')))
traceB = ts.tsygTrace(filename='../rbsp/data/trace.{}.B.dat'.format(date.strftime('%Y%m%d')))
```

```python
def mapBackground():
    lon0 = 270
    llcrnrlon = lon0 - 25.
    urcrnrlon = lon0 + 91.
    myMap = Basemap(projection='stere', lat_0=90, lon_0=lon0, resolution='l', 
                    llcrnrlat=20, llcrnrlon=llcrnrlon,
                    urcrnrlat=60, urcrnrlon=urcrnrlon, ax=gca())
    myMap.drawcoastlines(color='.5')
    myMap.fillcontinents(color='.8')
    # draw parallels and meridians.
    myMap.drawparallels(np.arange(-80.,81.,20.), color='.6')
    myMap.drawmeridians(np.arange(-180.,181.,20.), color='.6')
    myMap.drawmapboundary()
    
    return myMap
```

```python
def mhoFov(myMap, elev, dist, azim=[-90.,15.]):
    # MHO location
    mhpos = [42.6, 288.5]
    x0, y0 = myMap(mhpos[1], mhpos[0])
    myMap.scatter(x0, y0, s=20, zorder=5, c='k')
    text(x0*1.04, y0*0.96, 'MHO')
    
    # MHO fov
    nazims = 100
    azims = linspace(azim[0], azim[1], nazims)
    mhfov = zeros((nazims, 3))
    Rav = 6370.
    for iaz, taz in enumerate(azims):
        dd = gp.calcDistPnt(mhpos[0], mhpos[1], 0., dist=dist, el=elev, az=taz)
        mhfov[iaz,0] = dd['distLat']
        mhfov[iaz,1] = dd['distLon']
        mhfov[iaz,2] = dd['distAlt']
    x, y = myMap(mhfov[:,1], mhfov[:,0])
    myMap.plot(x, y, 'g')
    myMap.plot([x0,x[0]], [y0,y[0]], 'g')
    myMap.plot([x0,x[-1]], [y0,y[-1]], 'g')
    
    return mhfov
```

```python
def mhoSdRbspMap(elev, dist):
    # Map background
    myMap = mapBackground()
    figW = myMap.urcrnrx - myMap.llcrnrx
    figH = myMap.urcrnry - myMap.llcrnry
    
    # MHO fov
    mhfov = mhoFov(myMap, elev, dist)
    text(myMap.llcrnrx+.03*figW, myMap.llcrnry+.03*figH, 
            'Altitude (at max. range): {:3.0f}km'.format(mean(mhfov[:,2])), 
            bbox=dict(facecolor='w', alpha=0.5))
    title(r'MHO f-o-v: range={:5.0f} km, elev={} $^\circ$'.format(dist, elev))
    
    # Draw FP
    fovCol = (.5,.6,.6)
    fovAlpha = .2
    x, y = myMap(traceA.lonNH, traceA.latNH)
    myMap.scatter(x, y, color='r', zorder=2, s=5)
    x, y = myMap(traceB.lonNH, traceB.latNH)
    myMap.scatter(x, y, color='b', zorder=2, s=5)
    overlayFov(myMap, codes=['fhe','bks','wal'], hemi='north', maxGate=75, dateTime=traceA.datetime[0], 
        fovColor=fovCol, lineColor=fovCol, lineWidth=0, fovAlpha=fovAlpha)
    
    # More stuff
    rcParams.update({'font.size': 12})
    Title = '{}'.format(date.strftime('%Y - %b - %d'))
    text(myMap.llcrnrx, myMap.urcrnry, 
        Title, rotation=90, va='top', ha='right')
    text(myMap.llcrnrx, myMap.llcrnry, 
        'RBSP-A', color='r', va='top')
    text(myMap.llcrnrx + .2*figW, myMap.llcrnry, 
        'RBSP-B', color='b', va='top')
```
