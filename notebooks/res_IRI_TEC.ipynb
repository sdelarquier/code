```python
import sys
sys.path.append('/davitpy')
from models import iri
from datetime import datetime
from mpl_toolkits.basemap import Basemap
```

```python
# Inputs
dtime = datetime(2008, 3, 8, 8) # date and time UT
jf = np.ones(50)
jf[3:5] = 0 
jf[20] = 0 
jf[22] = 0 
jf[27:29] = 0 
jf[32] = 0 
jf[34] = 0 
jmag = 0.
alati = 40. 
along = -80.
iyyyy = dtime.year
mmdd = dtime.month*100 + dtime.day 
dhour = dtime.hour 
vbeg, vend, vstp = 10., 90., 1. 
lats = arange(vbeg, vend+vstp/10., vstp)
lons = linspace(-150,-30,len(lats))
oarr = np.zeros((100,1000))
tec = zeros((len(lons), len(lats)))
# Call fortran subroutine
for ilon, along in enumerate(lons):
    outf,oarr = iri.iri_web( jmag, jf, alati, along, iyyyy, mmdd, 1, dhour,
                            300., 600., 2, vbeg, vend, vstp, oarr)
    tec[ilon, :] = oarr[36,0:len(lats)]*1e-16
```

```python
fig = figure(figsize=(8,8))
ax = fig.add_subplot(111)
myMap = Basemap(projection='stere',lon_0=270, lat_0=45, resolution='l', 
                llcrnrlat=10, llcrnrlon=-120, 
                urcrnrlat=60, urcrnrlon=0.)
myMap.drawcoastlines(color='.5')
# draw parallels and meridians.
myMap.drawparallels(np.arange(-80.,81.,20.), color='.6')
myMap.drawmeridians(np.arange(-180.,181.,20.), color='.6')
myMap.drawmapboundary()
myMap.nightshade(dtime)

# Add data
x, y = myMap(*meshgrid(lons, lats))
cs = myMap.pcolormesh(transpose(x), transpose(y), tec, vmin=0, vmax=4)

# Add CB
cbar = myMap.colorbar(cs, location='bottom', pad="5%")
cbar.set_label(r'TEC [$\times 10^{16} \rm{m}^{-2}$]')

# Add title
title('{}'.format(dtime.strftime('%Y-%m-%d %H UT')))
fig.savefig('/home/sebastien/Desktop/iri_tec_evanomaly.pdf')
```
