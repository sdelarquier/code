##Van Allen Probes foot-points

* The following code gives you access to RBSP foot-points calculated at Virginia Tech using the Tsyganenko T96 magnetic field model with default quiet geomagnetic conditions (Dst Index : -5 nT, solar wind dynamic pressure : 2 nPa, IMF-By : 0 nT, IMF-Bz : -5 nT). The Python wrapper for T96 can be found [here](https://github.com/sdelarquier/tsyganenko).

* Ephemeris are obtained from JHU/APL ([athena.jhuapl.edu/](http://athena.jhuapl.edu/)). Note that ephemeris are degraded to a single decimal digit due to NASA policy. However, this does not seem to affect the footpoint calculations in any significant way.

```python
import rbsp
from datetime import datetime
```

```python
# Start of period of interests
sTime = datetime(2013,2,1,0)
# End of period (do not request more than 24 hours at once)
eTime = datetime(2013,2,1,6)
```

```python
# Some information about the required inputs
rbsp.rbspFp?
```

```python
# Get foot-points
fps = rbsp.rbspFp(sTime, eTime)
```

```python
# Get some information about what you just loaded
print fps
```
