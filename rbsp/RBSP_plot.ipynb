```python
import tsyganenko as ts
import numpy as np
from datetime import datetime
import sys
sys.path.append('/davitpy')
from pydarn.plot import overlayFov
from pydarn.radar import network
import matplotlib as mp
from mpl_toolkits.basemap import Basemap
```

```python
def plotOrbit(sdates, lats, lons, rhos=None):
    fig = figure(figsize=(25,4))
    ax = fig.add_subplot(131)
    plot(sdates, lats)
    xticks = mp.dates.HourLocator(interval=4)
    xticksfmt = mp.dates.DateFormatter('%H:%M')
    gca().xaxis.set_major_locator(xticks)
    gca().xaxis.set_major_formatter(xticksfmt)
    ylabel('latitude')
    ax = fig.add_subplot(132)
    plot(sdates, lons)
    xticks = mp.dates.HourLocator(interval=4)
    xticksfmt = mp.dates.DateFormatter('%H:%M')
    gca().xaxis.set_major_locator(xticks)
    gca().xaxis.set_major_formatter(xticksfmt)
    ylabel('longitude')
    if rhos is not None:
        ax = fig.add_subplot(133)
        plot(sdates, rhos)
        xticks = mp.dates.HourLocator(interval=4)
        xticksfmt = mp.dates.DateFormatter('%H:%M')
        gca().xaxis.set_major_locator(xticks)
        gca().xaxis.set_major_formatter(xticksfmt)
        ylabel('altitude')
```

```python
year, month, day, hour, minute, second, alt, lat, lon = np.genfromtxt('orbitposA.dat', unpack=True, skiprows=1)
dates = []
for y,m,d,h,mn,s in zip(year, month, day, hour, minute, second):
    dates.append( datetime( int(y),int(m),int(d),int(h),int(mn),int(s) ) )
dates = np.array(dates)
# Select a date
inds = [d.date()==date.date() for d in dates]
sdates = dates[np.array(inds)]
# trace points
lats = lat[np.array(inds)]
lons = lon[np.array(inds)]
rhos = alt[np.array(inds)]

plotOrbit(sdates, lats, lons, rhos)
```

```python
traceA = ts.tsygTrace(lats, lons, rhos, datetime=sdates)
```

```python
plotOrbit(traceA.datetime, traceA.latNH, traceA.lonNH)
```

```python
traceB = ts.tsygTrace(lats, lons, rhos, datetime=sdates)
```
